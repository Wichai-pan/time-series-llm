```python
import os
import re
```

```python
# Split a weekly routine file into separate daily routine files
def split_weekly_routine_by_day(input_file_path: str, output_dir: str):
    
    days_of_week = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]

    os.makedirs(output_dir, exist_ok=True)

    with open(input_file_path, "r", encoding="utf-8") as file:
        lines = file.readlines()

    base_name = os.path.splitext(os.path.basename(input_file_path))[0]

    current_day = None
    buffer = []

    for line in lines:
        day_match = next(
            (day for day in days_of_week if line.strip().startswith(f"{day}:")),
            None
        )

        if day_match:
            if current_day and buffer:
                output_path = os.path.join(
                    output_dir, f"{base_name}_{current_day}.txt"
                )
                with open(output_path, "w", encoding="utf-8") as out_file:
                    out_file.writelines(buffer)

            current_day = day_match
            buffer = [line]
        else:
            buffer.append(line)

    # Write final day
    if current_day and buffer:
        output_path = os.path.join(
            output_dir, f"{base_name}_{current_day}.txt"
        )
        with open(output_path, "w", encoding="utf-8") as out_file:
            out_file.writelines(buffer)

    print(f"[AgentSense] Split weekly routine into daily files → {output_dir.replace(os.sep, '/')}")
```

```python
# =========================
# Configuration (User-editable)
# =========================

# Weekly routine file generated in Step 3
# Expected format: {character_name}_routine_{env_id}.txt
WEEKLY_ROUTINE_FILE = "data/routine_data/Sarah_routine_env_0.txt"

# Output directory for per-day routine files
DAILY_ROUTINE_OUTPUT_DIR = "data/step_4_data"

# =========================
# Run
# =========================

print("[AgentSense] Splitting weekly routine into daily routines...")

split_weekly_routine_by_day(
    input_file_path=WEEKLY_ROUTINE_FILE,
    output_dir=DAILY_ROUTINE_OUTPUT_DIR
)

print("[AgentSense] Daily routine files generated successfully.")
```
