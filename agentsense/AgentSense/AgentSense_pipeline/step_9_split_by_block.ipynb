```python
import os
import re
```

```python
# Extracts the room name from a block by looking for the last valid room in parentheses
def extract_room_name_from_block(block):
    room_matches = re.findall(r'\(([^()]+)\)', block)
    for match in reversed(room_matches):
        if match.lower() in ["bedroom", "kitchen", "bathroom", "livingroom"]:
            return match.lower()
    return "unknown"
```

```python
def split_blocks_and_save(input_file_path, output_dir):
    
    if not os.path.exists(input_file_path):
        print(f"🚨 Input file not found: {input_file_path}")
        return
    
    os.makedirs(output_dir, exist_ok=True)
    base_name = os.path.splitext(os.path.basename(input_file_path))[0]
    
    with open(input_file_path, "r") as f:
        content = f.read()
    

    # Split the content into blocks separated by two newlines
    blocks = [block.strip() for block in content.strip().split('\n\n') if block.strip()]
    total_blocks = len(blocks)
    
    print(f"[AgentSense] Total blocks found: {total_blocks}")
    if total_blocks == 0:
        print("[AgentSense] No blocks detected. Check formatting (expected blank line between blocks).")
        return    
    
    previous_room = "bedroom"  # Always label the first file with bedroom

    for part_index, i in enumerate(range(0, total_blocks, 4), start=1):
        chunk = blocks[i:i+4]

        current_room = previous_room

        # Update previous_room for next round, only if we have 4 blocks
        if len(chunk) == 4:
            previous_room = extract_room_name_from_block(chunk[3])  # 4th block (index 3)

        # Build output filename and path
        output_filename = f"part_{part_index}_{base_name}_{current_room}.txt"
        output_path = os.path.join(output_dir, output_filename)

        with open(output_path, "w") as f:
            f.write("\n\n".join(chunk))

        print(f"[AgentSense] ✅ Saved: {output_path}")
```

```python
# =========================
# Configuration (User-editable)
# =========================

input_file_path = "data/step_8_data/label_regenerated_Sarah_routine_env_0_Monday_parsed_grounded.txt"
output_dir = "data/step_9_data"

split_blocks_and_save(input_file_path, output_dir=output_dir)
```
